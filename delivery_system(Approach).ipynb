```python
import folium
from geopy.distance import geodesic
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent='distance-tracker')


eiffel_tower_address = 'Eiffel Tower, Paris, France'
eiffel_tower_location = geolocator.geocode(eiffel_tower_address)


eiffel_tower_latitude = eiffel_tower_location.latitude
eiffel_tower_longitude = eiffel_tower_location.longitude


colosseum_address = 'Colosseum, Rome, Italy'
colosseum_location = geolocator.geocode(colosseum_address)

colosseum_latitude = colosseum_location.latitude
colosseum_longitude = colosseum_location.longitude


eiffel_tower_location_tuple = (eiffel_tower_latitude, eiffel_tower_longitude)
colosseum_location_tuple = (colosseum_latitude, colosseum_longitude)
distance = geodesic(eiffel_tower_location_tuple, colosseum_location_tuple).km

map_location = [eiffel_tower_latitude, eiffel_tower_longitude]
m = folium.Map(location=map_location, zoom_start=5)


folium.Marker(
    location=map_location,
    popup='Eiffel Tower',
    icon=folium.Icon(color='red')
).add_to(m)


colosseum_location = [colosseum_latitude, colosseum_longitude]
folium.Marker(
    location=colosseum_location,
    popup='Colosseum',
    icon=folium.Icon(color='green')
).add_to(m)


folium.PolyLine(
    locations=[map_location, colosseum_location],
    color='blue'
).add_to(m)


folium.Marker(
    location=[(eiffel_tower_latitude+colosseum_latitude)/2, (eiffel_tower_longitude+colosseum_longitude)/2],
    popup=f'Distance: {distance:.2f} km',
    icon=folium.Icon(color='purple')
).add_to(m)


m
```

```python
import folium
from geopy.distance import geodesic
from geopy.geocoders import Nominatim


geolocator = Nominatim(user_agent='delivery-tracker')


store_address = 'Connaught Place, New Delhi, India'
store_location = geolocator.geocode(store_address)


store_latitude = store_location.latitude
store_longitude = store_location.longitude


delivery_address = 'Pragati Maidan, New Delhi, India'
delivery_location = geolocator.geocode(delivery_address)


delivery_latitude = delivery_location.latitude
delivery_longitude = delivery_location.longitude


store_location_tuple = (store_latitude, store_longitude)
delivery_location_tuple = (delivery_latitude, delivery_longitude)
distance_km = geodesic(store_location_tuple, delivery_location_tuple).km

delivery_cost = distance_km * 20


map_location = [store_latitude, store_longitude]
m = folium.Map(location=map_location, zoom_start=14)


folium.Marker(
    location=map_location,
    popup='Connaught Place',
    icon=folium.Icon(color='green')
).add_to(m)

delivery_location = [delivery_latitude, delivery_longitude]
folium.Marker(
    location=delivery_location,
    popup='Chandni Chowk',
    icon=folium.Icon(color='red')
).add_to(m)

folium.PolyLine(
    locations=[map_location, delivery_location],
    color='blue'
).add_to(m)

distance_text = f'Distance: {distance_km:.2f} km'
delivery_cost_text = f'Delivery Cost: Rs {delivery_cost:.2f}'
in_transit_text = 'In Transit'
folium.Marker(
    location=[(store_latitude+delivery_latitude)/2, (store_longitude+delivery_longitude)/2],
    popup=f'{distance_text}<br>{in_transit_text}<br>{delivery_cost_text}',
    icon=folium.Icon(color='gray')
).add_to(m)

m
```

```python
import folium
import requests
from folium.features import CustomIcon


store_coords = (28.6328, 77.2197)
delivery_coords = (28.6139, 77.2495)

osrm_url = f'http://router.project-osrm.org/route/v1/driving/{store_coords[1]},{store_coords[0]};{delivery_coords[1]},{delivery_coords[0]}?geometries=geojson'
response = requests.get(osrm_url)


route_geometry = response.json()['routes'][0]['geometry']


m = folium.Map(location=store_coords, zoom_start=14)


folium.Marker(
    location=store_coords,
    popup='Connaught Place',
    icon=folium.Icon(color='red')
).add_to(m)

folium.Marker(
    location=delivery_coords,
    popup='Pragati Maidan',
    icon=folium.Icon(color='green')
).add_to(m)

folium.GeoJson(
    data=route_geometry,
    style_function=lambda x: {'color': 'blue', 'weight': 5}
).add_to(m)


alt_routes_url = f'http://router.project-osrm.org/route/v1/driving/{store_coords[1]},{store_coords[0]};{delivery_coords[1]},{delivery_coords[0]}?geometries=geojson&alternatives=true'

response = requests.get(alt_routes_url)



icon_coords = (28.6250, 77.2350)


icon_image = CustomIcon(
    '/Users/dhruvsood/Downloads/dbms 3/mapss/delivery-man.png',
    icon_size=(50, 50)
)


folium.Marker(
    location=icon_coords,
    icon=icon_image
).add_to(m)


m
```
